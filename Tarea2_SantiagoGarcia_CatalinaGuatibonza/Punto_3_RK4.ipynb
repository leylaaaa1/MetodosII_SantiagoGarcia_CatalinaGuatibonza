```python
import numpy as np
import matplotlib.pyplot as plt

condicion_in = [np.sqrt(2), 0]
ejex_size = 10
h = 0.01  

def Riccati(x, y):
    
    return x * y**2 - (2*y)/x - 1/x**3


def Runge_kutta_4(fun, x0, y0, ejex_size, h):
    
    x = condicion_in[0]
    y = condicion_in[1]
    x_valor = [x0]
    y_valor = [y0]
    
    while x < ejex_size:
        k1 = h * fun(x, y)
        k2 = h * fun(x + h/2, y + k1/2)
        k3 = h * fun(x + h/2, y + k2/2)
        k4 = h * fun(x + h, y + k3)
        
        x += h
        y += (k1 + 2*k2 + 2*k3 + k4) / 6
        
        
        x_valor.append(x)
        y_valor.append(y)
    
    return np.array(x_valor), np.array(y_valor)


x_valor, y_valor = Runge_kutta_4(Riccati, x0, y0, ejex_size, h)


plt.plot(x_valor, y_valor)
plt.xlabel('x')
plt.ylabel('y')
plt.title('SoluciÃ³n a Riccati con Runge-Kutta de cuarto orden')
plt.grid(True)
plt.show()
```
