```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la función que representa la ecuación diferencial
def du_dt(u, q):
    return u**q
```

```python
# Método de Euler para resolver la ecuación diferencial
def euler_method(q, u0, t0, tf, dt):
    t = np.arange(t0, tf, dt)
    u = np.zeros(len(t))
    u[0] = u0
    
    for i in range(1, len(t)):
        u[i] = u[i-1] + du_dt(u[i-1], q) * dt
    
    return t, u
```

```python
# Parámetros iniciales
u0 = 1  # Condición inicial
t0 = 0  # Tiempo inicial
tf = 10  # Tiempo final
dt = 0.01  # Paso de tiempo

# Valores de q para los cuales resolveremos la ecuación
qs = [0, 0.2, 0.4, 0.7, 0.9, 1.0]
```

```python
# Solución analítica para q < 1
def analytical_solution(t, q):
    return ((t * (1 - q) + 1) ** (1 / (1 - q))) if q != 1 else np.exp(t)

# Resolvemos la ecuación para cada valor de q y graficamos
plt.figure(figsize=(10, 6))

for q in qs:
    t, u_euler = euler_method(q, u0, t0, tf, dt)
    u_analytical = analytical_solution(t, q)
    
    plt.plot(t, u_euler, label=f'Euler q={q}')
    plt.plot(t, u_analytical, '--', label=f'Analytical q={q}')

plt.xlabel('t')
plt.ylabel('u(t)')
plt.title('Numerical solution of du/dt = u^q')
plt.legend()
plt.grid(True)
plt.show()
```
