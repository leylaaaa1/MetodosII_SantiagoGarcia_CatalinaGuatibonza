```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dudt(u, q):
    return u**q
```

```python
# Método de Euler para resolver la ecuación diferencial
def EulerMethod(q, u0, t0, tf, dt):
    t = np.arange(t0, tf, dt)
    h = t[1] - t[0]
    u = np.zeros(len(t))
    u[0] = u0
    
    for i in range(1, len(t)):
        u[i] = u[i-1] + dudt(u[i-1], q) * h
    
    return t, u
```

```python
u0 = 1
t0 = 0
tf = 10
dt = 0.01

qv = [0, 0.2, 0.4, 0.7, 0.9, 1.0] # Valores de q
```

```python
for q in qv:
    t, q_euler = EulerMethod(q, u0, t0, tf, dt)
    
    plt.plot(t, q_euler, label=f'q={q}')

plt.xlabel('t')
plt.ylabel('u(t)')
plt.title('Numerical solution of du/dt = u^q')
plt.legend()
plt.show()
```
