```python
import numpy as np
```

```python
t = np.linspace(0,10,100)
h = 0.1
```

```python
def CalcularJ(h):

    dxn1_dxn = 1
    dvn1_dxn = 0
    dvn1_dvn = 1
    dxn1_dvn = h
    
    J = dxn1_dxn * dvn1_dvn - dxn1_dvn * dvn1_dxn
    
    return J

Jacob = CalcularJ(h)
```

```python
print("El Jacobiano es igual a:", Jacob)
```
