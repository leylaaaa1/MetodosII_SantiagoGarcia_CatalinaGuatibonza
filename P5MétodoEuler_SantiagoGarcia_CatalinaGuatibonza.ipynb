```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros iniciales
alpha = -1
u0 = 1
t_final = 20  # Tiempo final para la simulación
delta_ts = [1.1, 1.5, 1.9]  # Valores de Δt a simular
```

```python
# Solución exacta para comparación
t_exact = np.linspace(0, t_final, 1000)
u_exact = np.exp(alpha * t_exact)
plt.figure(figsize=(10, 5))

# Graficar la solución exacta
plt.plot(t_exact, u_exact, 'k', label='Solución exacta $u(t) = e^{\\alpha t}$')

for delta_t in delta_ts:
    N = int(t_final / delta_t)  # Número de pasos
    t = np.linspace(0, t_final, N+1)
    u = np.zeros(N+1)
    u[0] = u0
    
    # Método de Euler
    for i in range(N):
        u[i+1] = u[i] + alpha * u[i] * delta_t

    # Graficar la solución numérica para cada Δt
    plt.plot(t, u, '-o', label=f'Solución numérica con Δt = {delta_t}')
    
plt.title('Comparación de Soluciones Numéricas y Exactas de la Ecuación Diferencial')
plt.xlabel('Tiempo (s)')
plt.ylabel('u(t)')
plt.legend()
plt.grid(True)
plt.show()
```
